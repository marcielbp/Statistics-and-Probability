```python
import pandas as pd
import numpy as np
import folium
import folium.plugins as plugins
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
import time
import os
import re
import fileinput
```

```python
fp = "SHAPES/municipios_simp.json"
cidadesBR = pd.read_csv("csv/muniBR-ALL.csv",delimiter=',')
dataSet = pd.read_csv("csv/OB_DATA_2016.csv",sep=",")
dataFrameBase = pd.DataFrame(dataSet, columns = ['CODMUNRES','IDADE','CAUSABAS'])
```

```python
onlyAsma = dataFrameBase[dataFrameBase['CAUSABAS'].str.contains('J1')]#doencas respiratorias
```

```python
otherDisease = dataFrameBase[dataFrameBase['CAUSABAS'].str.contains('J2')]#doencas respiratorias
```

```python
data = gpd.read_file(fp)
```

```python
data
```

```python
data['OB_ASMA']=0
data['OB_RESP']=0
data['POPULACAO']=10000 #debug
data['OB_ASMA_POP']=0
data['OB_RESP_POP']=0
```

```python
numCities = len(cidadesBR)
```

```python
for i in range(0,numCities):
	codMun=cidadesBR.iloc[i, 1]
	popMun=cidadesBR.iloc[i, 3]
	data.loc[data.GEOCODIGO.str.contains(str(codMun)), 'POPULACAO'] =popMun

values = onlyAsma['CODMUNRES'].value_counts().keys().tolist()
counts = onlyAsma['CODMUNRES'].value_counts().tolist()
numCounts = len(counts)
```

```python
mapa = folium.Map(location=[-15.788497,-47.879873],tiles='Mapbox Bright', zoom_start=4)

```

```python
for i in range(0,numCounts):
	data.OB_ASMA[data.GEOCODIGO.str.contains(str(values[i]))]=counts[i]
	#folium.Marker(location=[-15.788497,-47.879873], popup=Popup(counts[i])).add_to(mapa)
data['OB_ASMA_POP']=100000*data['OB_ASMA']/data['POPULACAO']

```

```python
valuesD = otherDisease['CODMUNRES'].value_counts().keys().tolist()
countsD = otherDisease['CODMUNRES'].value_counts().tolist()
numCountsD = len(countsD)
```

```python
for i in range(0,numCountsD):
    data.OB_RESP[data.GEOCODIGO.str.contains(str(valuesD[i]))]=countsD[i]
    #folium.Marker(location=[-15.788497,-47.879873], popup=Popup(counts[i])).add_to(mapa)
data['OB_RESP_POP']=100000*data['OB_RESP']/data['POPULACAO']
```

```python
fmap= folium.Map(location=[-20,-40], 
                    zoom_start=5,
                    tiles='cartodbdark_matter',
                    control_scale=True)
```

```python
g1 = folium.Choropleth(
    geo_data=fp,
    name='choropleth',
    data=data,
    columns=['GEOCODIGO', 'OB_ASMA_POP'],
    key_on='feature.properties.GEOCODIGO',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0,
    bins=6,
    smooth_factor=2,
    legend_name='Obitos/100000 hab por Pneumonia em 2016',
    highlight=False
).add_to(fmap)
```

```python

g1 = folium.GeoJson(
        data[['NOME', 'geometry','OB_ASMA_POP']].to_json(), 
        name='Numero de Obitos por Asma',
        show=True,
        style_function=lambda feature: {
            'fillColor': 'green',
            'color': 'white',
            'weight': 0,
            'fillOpacity':0
        },   
        highlight_function=lambda x: {'weight':0.1, 
                                      'color':'black',
                                      'fillOpacity':0.2},    
        tooltip=folium.features.GeoJsonTooltip(
            fields=['NOME','OB_ASMA_POP'],
            aliases=['Cidade:','Obitos'],                              
        ),  
).add_to(fmap)
```

```python
g2 = folium.Choropleth(
    geo_data=fp,
    name='choropleth',
    data=data,
    columns=['GEOCODIGO', 'OB_RESP_POP'],
    key_on='feature.properties.GEOCODIGO',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0,
    bins=6,
    smooth_factor=2,
    legend_name='Obitos/100000 hab por doencas respiratorias em 2016',
    highlight=False
).add_to(fmap)
```

```python


g2 = folium.GeoJson(
        data[['NOME', 'geometry','OB_RESP_POP']].to_json(), 
        name='Numero de Obitos por Asma',
        show=True,
        style_function=lambda feature: {
            'fillColor': 'green',
            'color': 'white',
            'weight': 0,
            'fillOpacity':0.1
        },   
        highlight_function=lambda x: {'weight':0.3, 
                                      'color':'black',
                                      'fillOpacity':0.4},    
        tooltip=folium.features.GeoJsonTooltip(
            fields=['NOME','OB_RESP_POP'],
            aliases=['Cidade:','Obitos'],                              
        ),  
).add_to(fmap)
```

```python
from folium.plugins import Search
citysearch = Search(
    layer=g1,
    geom_type='Polygon',
    placeholder='Busca Cidade',
    collapsed=True,
    search_label='NOME'
).add_to(fmap)
```

```python
folium.LayerControl(collapsed=False).add_to(fmap)
fmap.save("mapaT.html") 
```
