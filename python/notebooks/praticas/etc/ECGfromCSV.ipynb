```python
freq = 250 #1kHz
f_s = 250  # Sampling rate, or number of measurements per second

a250 = pd.read_csv('download_2020-3-13_19-44-55_8ms.csv')


## Matplotlib Method

from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Norasi']
rcParams['font.size'] = 12
#plt.rcParams['font.family'] = 'sans-serif'
#plt.rcParams['font.sans-serif'] = 'Times New Roman'
fig, ax = plt.subplots(figsize=(14, 3.5))
ax.plot(a250['t'], a250['y'],color='#900C3F',linewidth=2)
ax.set_title('ECG collected Signal',fontsize=12)
ax.set_ylabel('Amplitude (normalized mV)')
ax.set_xlabel('Time (sec)')
ax.grid(b=True,which='major', axis='x',color='k', linestyle='-', linewidth=1)
ax.grid(which='major', axis='y',color='k', linestyle='-', linewidth=0.25)
ax.minorticks_on()
ax.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.25)
ax.set_xlim([0,5])
ax.set_ylim([-10,80])
fig.savefig('ecgFig.pdf')
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```
